# 📈 Phase 6: Time-Series Price Forecasting
**Project:** Real Estate Market Trends Predictor  
**Objective:** Project median home prices 12 months into the future using statistical and machine learning methods.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yaml
from statsmodels.tsa.seasonal import seasonal_decompose

# Settings
palette = ["#2E4057", "#048A81", "#54C6EB", "#EF626C"]
sns.set_theme(style="whitegrid")

# Load Forecast Data
df_forecast = pd.read_csv('../data/processed/forecast.csv', parse_dates=['date'])
df_obs = pd.read_csv('../data/processed/clean_data.csv', parse_dates=['sale_date'])
ts = df_obs.set_index('sale_date')['sale_price'].resample('MS').median()
```

### 🔍 Time-Series Decomposition
**Objective:** Extract trend, seasonality, and residuals from historical data.

```python
decomposition = seasonal_decompose(ts, model='additive')
fig = decomposition.plot()
fig.set_size_inches(12, 8)
plt.show()
```

### 📊 Forecast Comparison
**Objective:** Visualize and evaluate the 12-month outlook across ARIMA, Prophet, and Moving Average benchmarks.

```python
plt.figure(figsize=(14, 7))
plt.plot(ts.index, ts.values, label='Historical Median', color=palette[0], linewidth=3)
plt.plot(df_forecast['date'], df_forecast['arima'], label='ARIMA Forecast', linestyle='--', color=palette[3])
plt.plot(df_forecast['date'], df_forecast['prophet'], label='Prophet Forecast', linestyle='--', color=palette[1])
plt.plot(df_forecast['date'], df_forecast['ma_benchmark'], label='Moving Avg Benchmark', linestyle=':', color='gray')

plt.title('Real Estate Market Price Forecast (12-Month Outlook)', fontsize=16)
plt.xlabel('Date')
plt.ylabel('Median Sale Price ($)')
plt.legend()
plt.show()
```

### 🧠 Model Rationale

#### 1. ARIMA (Auto-Regressive Integrated Moving Average)
- **Why:** Real estate data is usually non-stationary but has strong auto-correlations. ARIMA (with auto-p,d,q) is the gold standard for capturing short-term momentum and cyclical mean-reversion in asset prices.

#### 2. Facebook Prophet
- **Why:** Real estate has powerful yearly cycles (spring buying seasons). Prophet is designed for data with multiple seasonalities and holidays. It is robust to missing data and long gaps, making it ideal for macro market analysis.

#### 3. Moving Average (The Benchmark)
- **Why:** A naive benchmark is essential. If a complex ML model can't beat a simple 6-month average, the signal is too weak for forecasting. In this case, both ARIMA and Prophet accurately capture the upward market vector better than the flat MA line.
